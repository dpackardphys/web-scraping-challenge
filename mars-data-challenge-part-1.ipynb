```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://redplanetscience.com/"
browser.visit(url)
```

```python
# url = "https://redplanetscience.com/"
# response = requests.get(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("div", class_="list_text")
print(results)
```

```python
# results = soup.find_all('div', class_='list-text')
# results2 = results1.find_all('div', class_='row')
# title_results = results2.find_all('div', class_='content-title')
# preview_results = results2.find_all('div', class_='article-teaser-body')

articles_list = []

# loop over results to get article data
for result in results:
    # scrape the article title 
    title = result.find('div', class_='content_title').text
    
    # scrape the article teaser
    preview = result.find('div', class_='article_teaser_body').text
    
    # print article data
    print('-----------------')
    print(title)
    print(preview)


    # Dictionary to be inserted into list
    article = {
        'title': title,
        'preview': preview
    }

    # Insert dictionary into list
    articles_list.append(article)
```
